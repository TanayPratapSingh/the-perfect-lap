```python
import pandas as pd

# Load full circuits + races
circuits = pd.read_csv("/workspaces/group-project-team-omega/data/circuits.csv")
races = pd.read_csv("/workspaces/group-project-team-omega/data/races.csv")

# Step 1: Filter races from year >= 2000
races_2000 = races[races["year"] >= 2000]

# Step 2: Extract unique circuitIds used after 2000
used_circuits = races_2000["circuitId"].unique()

print("Total circuits used after 2000:", len(used_circuits))
print("List of circuitIds:", used_circuits)

# Step 3: Filter circuits.csv to only those used
circuits_filtered = circuits[circuits["circuitId"].isin(used_circuits)]

# Step 4: Save to new CSV
circuits_filtered.to_csv("circuits_2000plus.csv", index=False)

print("Saved: csv/circuits_2000plus.csv")
```
