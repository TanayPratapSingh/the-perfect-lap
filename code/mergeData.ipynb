```python
import pandas as pd
import os

# Make sure folder exists
os.makedirs("cleanedData", exist_ok=True)

# Load cleaned datasets
results = pd.read_csv("cleanedData/clean_results.csv")
races = pd.read_csv("cleanedData/clean_races.csv")
drivers = pd.read_csv("cleanedData/clean_drivers.csv")
constructors = pd.read_csv("cleanedData/clean_constructors.csv")

# ---------- STEP 1: Merge results + races ----------
df = results.merge(races, on="raceId", how="left")

# ---------- STEP 2: Merge drivers ----------
df = df.merge(drivers, on="driverId", how="left")

# ---------- STEP 3: Merge constructors ----------
df = df.merge(constructors, on="constructorId", how="left")

# ---------- SAVE FINAL MASTER ----------
df.to_csv("cleanedData/master_planA.csv", index=False)

print("Master dataset created → CleanedData/master_planA.csv")
```
