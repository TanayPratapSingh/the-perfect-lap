```python
import pandas as pd

# Load your imputed dataset
df = pd.read_csv("cleanedData/new_master_planA_imputed.csv")  
# change name if needed

print("Before filtering:", len(df))

# ----------------------------------------------------
# 1. Remove weird pit laps (<1 or >40)
# ----------------------------------------------------
df = df[(df["first_pit_lap"] >= 1) & (df["first_pit_lap"] <= 40)]


print("After filtering weird pits:", len(df))

# ----------------------------------------------------
# 2. Save final dataset for all models
# ----------------------------------------------------
df.to_csv("cleanedData/Final_Model_data.csv", index=False)

print("Saved Final_Model_data.csv")
print("Total rows:", len(df))
print("Columns:", df.columns.tolist())
```
